The *central limit theorem* states that when independent random variables are added together, they sum to a normal distribution even if the original variables themselves are not normally distributed. For your homework test this assumption.

3. Using the *runif* function, run three experiments by simulating the outcomes of rolling a single, six-sided die. Show the distribution of each experiment. Show how the simulated means compare to the expected mean of a fair roll.
        	a) Exp 1: 10 independent throws
        	b) Exp 2: 1,000 independent throws
        	c) Exp 3: 10,000 independent throws


```r
# -------------------------------
# Question 3
# -------------------------------
library(ggplot2)

a = 0
b = 6
n1 = 10
n2 = 1000
n3 = 10000
y1 = 1.66
y2 = 166.66
y3 = 1666.66
dice <- data.frame(length=ceiling(runif(n1,min=a,max=b)))
ggplot(dice,aes(length))+
  geom_histogram(fill='white',color='black',binwidth = 1)+
  geom_hline(yintercept=y1)
dice2 <- data.frame(length=ceiling(runif(n2,min=a,max=b)))
ggplot(dice2,aes(length))+
  geom_histogram(fill='white',color='black',binwidth = 1)+
  geom_hline(yintercept=y2)
dice3 <- data.frame(length=ceiling(runif(n3,min=a,max=b)))
ggplot(dice3,aes(length))+
  geom_histogram(fill='white',color='black',binwidth = 1)+
  geom_hline(yintercept=y3)

```

4. Instead of rolling one die, run a set of experiments reporting the outcomes of rolling multiple dice at the same time.
        	a) Exp 1: 10,000 throws, 1 die
        	b) Exp 2: 10,000 throws, 2 dice
        	c) Exp 3: 10,000 throws, 3 dice
        	d) Exp 4; 10,000 throws, 6 dice.

Show the distribution of results for each experiment. Which of these experiments produces a distribution most similar to a normal normal distribution? Justify your conclusion using Q-Q plots.

**Answer:** 

```r
# -------------------------------
# Question 4
# -------------------------------

exp1 <-data.frame(die=ceiling(runif(n3,min=a,max=b)))
ggplot(exp1,aes(die))+
  geom_histogram(fill='white',color='black',binwidth = 1)+
    ggtitle("Exp1 1 die")
exp2 <-data.frame(die=ceiling(runif(n3,min=a,max=b)+runif(n3,min=a,max=b)))
ggplot(exp2,aes(die))+
  geom_histogram(fill='white',color='black',binwidth = 1)+
    ggtitle("Exp2 2 die")
exp3 <-data.frame(die=ceiling(runif(n3,min=a,max=b)+runif(n3,min=a,max=b)+runif(n3,min=a,max=b)))
ggplot(exp3,aes(die))+
  geom_histogram(fill='white',color='black',binwidth = 1)+
    ggtitle("Exp3 3 die")
exp4 <-data.frame(die=ceiling(runif(n3,min=a,max=b)+runif(n3,min=a,max=b)+runif(n3,min=a,max=b)+runif(n3,min=a,max=b)+runif(n3,min=a,max=b))+runif(n3,min=a,max=b))
ggplot(exp4,aes(die))+
  geom_histogram(fill='white',color='black',binwidth = 1)+
    ggtitle("Exp4 6 die")

ggplot(exp1,aes(sample=die))+geom_point(stat = 'qq') + ggtitle("exp 1 qq plot")   
ggplot(exp2,aes(sample=die))+geom_point(stat = 'qq') + ggtitle("exp 2 qq plot")   
ggplot(exp3,aes(sample=die))+geom_point(stat = 'qq') + ggtitle("exp 3 qq plot")   
ggplot(exp4,aes(sample=die))+geom_point(stat = 'qq') + ggtitle("exp 4 qq plot")   

```

Exp 4 has the most normal distribution seen by the connected and straight line in the qq plot matching up most closely to the theoretical normal distribution.
